```python
import yoda
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
from matplotlib.ticker import AutoMinorLocator
import os



#funo para fazer o grfico dos dados experimentais do ATLAS
def DataFromExperiment(file, scatter, min = 0, max = 0):
    '''
    Grab data from the Yoda.Scatter2D named scatter from the file yoda, following
    the method that colaborations (ALICE, ATLAS, CMS) save them.

    (str, str) -> (list(np.array))
    '''
    data = yoda.read(file)[scatter]

    if max == 0:
        max = len(data.xVals()) + 1

    x = np.asarray(data.xVals())[min:max]
    y = np.asarray(data.yVals())[min:max]
    err = np.asarray((data.yMaxs() - data.yMins()) / 2)[min:max]
    errx = np.asarray((data.xMaxs() - data.xMins()) / 2)[min:max]
    return [x, y, err, errx]






plt.style.use('plotconfig.mplstyle')


# Useful documentation:
# YODA: https://yoda.hepforge.org/pydoc/
# MATPLOTLIB: https://matplotlib.org/
# NUMPY: https://numpy.org/doc/

# Don't forget to load necessary packages in SAMPA
# source /cvmfs/alice.cern.ch/etc/login.sh
# eval `alienv printenv VO_ALICE@Rivet::2.7.2-alice2-1`




yodaPbPb = '/home/monalisa/Desktop/Mestrado/JetMass/Graph/pp_flow2.2_recoils_off_R0.4.yoda'
obs = '/JET_MASS/Jet_Mass_60_80'

    
    
histos_PbPb = yoda.read(yodaPbPb)
    
PbPb_jet = histos_PbPb[obs]

N_jets = PbPb_jet.sumW()

PbPb_jet.scaleW(1/N_jets)


x = np.asarray(PbPb_jet.xVals())
y = np.asarray(PbPb_jet.yVals())
yerr = np.asarray((PbPb_jet.yMaxs() - PbPb_jet.yMins()) / 2)
xerr = np.asarray((PbPb_jet.xMaxs() - PbPb_jet.xMins()) / 2)
                    
                

arr = [x, y, yerr, xerr]




#ax1 = plt.subplot(131)
fig, (axs1, axs2, axs3) = plt.subplots(ncols=3, nrows=1, sharex=False, sharey=True, gridspec_kw={'hspace': 0, 'wspace': 0}, figsize=(15, 5), dpi = 1500)




axs1.plot(arr[0], arr[1], color='r', linestyle='--', label='recoils off')



# Define plot
#plt.axhline(y=1.0, xmin=0., xmax=1200., alpha=0.5, color='k', linestyle='--')
#plt.figure()



#Plot dos dados experimentais            
ALICE1 = DataFromExperiment('HEPData-ins1512107-v1-Table_4.yoda', '/REF/ALICE_2017_I1512107/d04-x01-y01')

axs1.errorbar(*ALICE1, color='b', fmt='o', ms=5, elinewidth = 1., capsize = 5, capthick = 1, label='ALICE data')            
#antes fmt='s'            





axs1.set_xlabel(r'$M_{ch~jet}$' '  (GeV/$c^{2}$)',  loc = 'right', fontsize=18)
axs1.set_ylabel(r'$\frac{1}{N_{jets}}\frac{dN}{dM_{ch~jet}}$' '  ($c^{2}$/GeV)', loc='top',  fontsize=19)
axs1.text(8., 0.15, 'JEWEL 2.2 + PYTHIA + FLOW')
axs1.text(12., 0.10, '$0-10\%$ pp R = 0.4')
axs1.text(5., 0.17, '60 < $p_{T, ch~jet}}$ < 80 (GeV/c)')
#ax1.set_legend(loc=5)
axs1.set_xlim(0., 22.)
axs1.set_ylim(0., 0.21)
axs1.minorticks_on()




'''
Start the new plot for 80 < pt < 100
'''



obs2 = '/JET_MASS/Jet_Mass_80_100'

        
PbPb_jet2 = histos_PbPb[obs2]

N_jets2 = PbPb_jet2.sumW()

PbPb_jet2.scaleW(1/N_jets2)


x2 = np.asarray(PbPb_jet2.xVals())
y2 = np.asarray(PbPb_jet2.yVals())
yerr2 = np.asarray((PbPb_jet2.yMaxs() - PbPb_jet2.yMins()) / 2)
xerr2 = np.asarray((PbPb_jet2.xMaxs() - PbPb_jet2.xMins()) / 2)
                    
                

arr2 = [x2, y2, yerr2, xerr2]



axs2.plot(arr2[0], arr2[1], color='r', linestyle='--', label='recoils off')



#Plot dos dados experimentais            
ALICE2 = DataFromExperiment('HEPData-ins1512107-v1-Table_5.yoda', '/REF/ALICE_2017_I1512107/d05-x01-y01')

axs2.errorbar(*ALICE2, color='b', fmt='o', ms=5, elinewidth = 1., capsize = 5, capthick = 1, label='ALICE data')            
#antes fmt='s'  



axs2.set_xlabel(r'$M_{ch~jet}$' '  (GeV/$c^{2}$)', loc = 'right', fontsize=18)
#plt.ylabel(r'$\frac{1}{N_{jets}}\frac{dN}{dM_{ch~jet}}$' '  ($c^{2}$/GeV)', loc='top')
axs2.text(5., 0.17, '80 < $p_{T, ch~jet}}$ < 100 (GeV/c)')
axs2.text(12., 0.12, '$\sqrt{s_{NN}}=5.02$ TeV')
#axs2.set_legend(loc=5)
axs2.set_xlim(0., 22.)
axs2.set_ylim(0., 0.21)
axs2.minorticks_on()




'''
Start the new plot for 100 < pt < 120
'''



obs3 = '/JET_MASS/Jet_Mass_100_120'

        
PbPb_jet3 = histos_PbPb[obs3]

N_jets3 = PbPb_jet3.sumW()

PbPb_jet3.scaleW(1/N_jets3)


x3 = np.asarray(PbPb_jet3.xVals())
y3 = np.asarray(PbPb_jet3.yVals())
yerr3 = np.asarray((PbPb_jet3.yMaxs() - PbPb_jet3.yMins()) / 2)
xerr3 = np.asarray((PbPb_jet3.xMaxs() - PbPb_jet3.xMins()) / 2)
                    
                

arr3 = [x3, y3, yerr3, xerr3]
      
            
axs3.plot(arr3[0], arr3[1], color='r', linestyle='--', label='recoils off')


#Plot dos dados experimentais            
ALICE3 = DataFromExperiment('HEPData-ins1512107-v1-Table_6.yoda', '/REF/ALICE_2017_I1512107/d06-x01-y01')

axs3.errorbar(*ALICE3, color='b', fmt='o', ms=5, elinewidth = 1., capsize = 5, capthick = 1, label='ALICE data $\sqrt{s}$ = 2.76 TeV')            
#antes fmt='s'  




axs3.set_xlabel(r'$M_{ch~jet}$' '  (GeV/$c^{2}$)', loc = 'right',fontsize=18)
#axs3.set_ylabel(r'$\frac{1}{N_{jets}}\frac{dN}{dM_{ch~jet}}$' '  ($c^{2}$/GeV)', loc='top')
axs3.text(5., 0.17, '100 < $p_{T, ch~jet}}$ < 120 (GeV/c)')
plt.legend(loc=5)
axs3.set_xlim(0., 25.)
axs3.set_ylim(0., 0.21)
axs3.minorticks_on()



# Save
plt.savefig('pp_jetmass_5020_R0.4_jewel_2.2.0.pdf')
plt.savefig('pp_jetmass_5020_R0.4_jewel_2.2.0.png')


#Mostrar grfico
plt.show()
```
